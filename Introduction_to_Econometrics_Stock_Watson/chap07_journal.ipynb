Economic Journals Data Set

This dataset contains data on 180 economics journals for the year 2000.  
These data are provided by Professor Theodore Bergstrom of the Department of 
Economics at the University of California, San Diego.  
If you are interested in learning more about the economics of economics journals, 
see Bergstrom√ïs article in the Fall 2001 Journal of Economic Perspectives. 



```r
library(haven)
library(dplyr)
library(visreg)
journal <- read_dta("http://fmwww.bc.edu/ec-p/data/stockwatson/journals.dta")
```

```r
glimpse(journal)
```

```r
library(ggplot2)
p= ggplot(data = journal, mapping = aes(x=libprice  , y=citestot) )
p + geom_point()+ stat_smooth(method = "lm" , fill=NA)
```

```r

p= ggplot(data = journal, mapping = aes(x=libprice  , y=oclc, color = as.factor(field))) 
p+ geom_point()
```

```r
p= ggplot(data = journal, mapping = aes(x=libprice  , y=citestot, color = as.factor(society))) 
p+ geom_point()
```
