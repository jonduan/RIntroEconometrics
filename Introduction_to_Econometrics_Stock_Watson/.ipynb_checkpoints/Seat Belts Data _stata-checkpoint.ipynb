```python
%%stata 
*gen logincome = log(income)
reg fatalityrate  sb_useage speed65 speed70 ba08 drinkage21 logincome age, vce(cluster STATE) 
```

```python
%%stata 
*http://www.stata.com/support/faqs/data-management/creating-dummy-variables/
tabulate state, generate(stateB)

*Least squares with dummy variables (no constant term) .
 
reg fatalityrate  sb_useage speed65 speed70 ba08 drinkage21 logincome age stateB*,  vce(cluster STATE) nocons
```

```python
%%stata 
reg fatalityrate  sb_useage speed65 speed70 ba08 drinkage21 logincome age stateB*,  vce(cluster STATE) 
testparm stateB*
```

```python
%%stata 
reg fatalityrate  sb_useage speed65 speed70 ba08 drinkage21 logincome age i.STATE,  vce(cluster STATE) 
```

```python
%%stata 
testparm STATE
```

```python
%%stata 


*http://www.stata.com/support/faqs/data-management/creating-group-identifiers/

* egen STATE = group(state)

*Within estimation
xtreg fatalityrate  sb_useage speed65 speed70 ba08 drinkage21 logincome age, fe i(STATE) vce(cluster STATE)
```

To get two-way fixed effects in official Stata, one must add the year dummies. 


This is easily done in Stata 11 with an term like: -i.year- assuming that "year' is the variable name for your time variable. 

In Stata 8, a separate command like: -tabulate year, gen(yrdum)- will provide permanent dummies for time with "yrdum" being the stub name for each of the separate dummy variables for each of the periods in your data set.

```python
%%stata 


*http://www.stata.com/support/faqs/data-management/creating-group-identifiers/

* egen STATE = group(state)

*Within estimation
xtreg fatalityrate  sb_useage speed65 speed70 ba08 drinkage21 logincome age i(year), fe i(STATE) r
```

```python
%%stata 


*http://www.stata.com/support/faqs/data-management/creating-group-identifiers/

* egen STATE = group(state)

*Within estimation
xtreg fatalityrate  sb_useage speed65 speed70 ba08 drinkage21 logincome age, fe i(STATE) 
```
