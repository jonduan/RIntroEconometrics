```r
###########
# Sampling distribution of the sample mean of a Bernoulli rv 
###########
n<-10; p<-.78; reps<-20000
y=rbinom(reps,n,p)/n
hist(y,probability=TRUE,breaks=seq(0,1,by=1/n))
x=seq(0,1,by=.01)
lines(x,dnorm(x,p,sqrt(p*(1-p)/n)),col=2,lty=1,lwd=3)
```

```r
###########
# Function for the sampling distribution of the sample mean of a Bernoulli rv with sample size n and probability of success p
###########
# generates reps sample means and plots them on a histogram
sampledist<-function(n,p,reps=50){
# create a matrix of sample random variables
y=rbinom(reps,n,p)/n
# plot using a histogram
hist(y,probability=TRUE,breaks=seq(0,1,by=1/n))
# add the normal pdf for comparison
x=seq(0,1,by=.01)
lines(x,dnorm(x,p,sqrt(p*(1-p)/n)),col=2,lty=1,lwd=3)
}
```

```r
###########
## compare sampling distributions
###########
# observe the convergence of the distribution to a normal distribution as n gets larger
sampledist(10,.78)
sampledist(50,.78)
sampledist(250,.78)
```

```r

###########
# Normalized sampling distribution
###########
n<-10; p<-.78; reps<-20000
y=rbinom(reps,n,p)/n
hist((y-p)/sqrt(p*(1-p)/n),probability=TRUE,xlim=c(-3,3))
x=seq(-3,3,by=.01)
lines(x,dnorm(x,0,1),col=2,lty=1,lwd=3)
```

```r
###########
# Function for normalized sampling distribution
###########
# for this distribution the mean is zero and variance is 1
normsampledist<-function(n,p){
reps<-20000
y<-rbinom(reps,n,p)/n
z<-(y-p)/sqrt(p*(1-p)/n)
hist(z,probability=TRUE,xlim=c(-3,3))
x=seq(-3,3,by=.01)
lines(x,dnorm(x,0,1),col=2,lty=1,lwd=3)
}
```

```r

###########
# compare sampling distributions
###########
normsampledist(10,.78)
normsampledist(50,.78)
normsampledist(250,.78)
```
